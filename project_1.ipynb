```python
#✅ import  that library's

import pandas as pd
from faker import Faker
from tabulate import tabulate
import random
fake=Faker()
import sqlite3
```

```python
#✅ sql-connection in colab to sqlite3

class sql_connection():
  def __init__(self,name):
    self.name=name

  def connection(self): #✅ connection code
    mydb=sqlite3.connect(self.name)
    mycursor=mydb.cursor()
    mycursor.execute("PRAGMA foreign_keys = ON")
    print("Database created")
    return mycursor,mydb

students=1000
start=1001
name="project_1.db"
connect=sql_connection(name)
mycursor,mydb=connect.connection()
print(mycursor,mydb)
```

```python
#✅ create dataset in faker library useing and create table and insert that values
class students_table():
  def __init__(self,n,start,mycursor,mydb):
    self.n=n
    self.start=start
    self.mycursor=mycursor
    self.mydb=mydb
    self.data=[]

  def create_data_1(self): #✅ table needed values creation
    num =(range(self.start, self.start+self.n))
    for i in num:
      students_id=i
      name=fake.name()
      age=random.randint(21,35)
      gender="".join(random.choice(["male","female","other"]))
      email=fake.email()
      ph_no=fake.phone_number()
      enrollment_year=random.randint(2024,2025)
      course_batch="abc"+str(random.randint(1,10))
      city=fake.city()
      graduation_year=random.randint(2010,2024)
      self.data.append([students_id,name,age,gender,email,ph_no,enrollment_year,course_batch,city,graduation_year])
    print("data created")
    return self.data

  def create_table_sql_1(self): #✅ table creation in sql
     self.mycursor.execute('''create table students_table ( student_id integer primary key,name text,age integer,gender text,
            email text,phone_no text,enrollment_year integer,course_batch text,city text,graduation_year int ) ''')
     print("table created")


  def values_inserting_1(self): #✅ values inserting list of tuples method
    frame=("insert into students_table values (?,?,?,?,?,?,?,?,?,?)")
    self.mycursor.executemany(frame,self.data)
    self.mydb.commit()
    print("values inserted")

students_data=students_table(students,start,mycursor,mydb) #✅ call the class
students_data.create_data_1()
students_data.create_table_sql_1()
students_data.values_inserting_1()
```

```python
%%writefile app.py
import streamlit as st
import sqlite3
import pandas as pd

class ConnectDB:
    def __init__(self, name='project_1.db'):
        self.pd = pd
        self.conn = sqlite3.connect(name)
        self.mycursor = self.conn.cursor()

    def execute_query(self, query, params=None):
        if params:
            self.mycursor.execute(query, params)
        else:
            self.mycursor.execute(query)
        out = self.mycursor.fetchall()
        df = self.pd.DataFrame(out, columns=[i[0] for i in self.mycursor.description])
        return df

    def save_dataframe(self, df, table_name):
        df.to_sql(table_name, self.conn, index=False, if_exists='replace')

class StreamlitPage:
    def __init__(self, pd, st):
        self.pd = pd
        self.st = st
        self.mydb = ConnectDB()
        self.page = self.st.sidebar.radio(
            "✳️ Pages",
            ["🏠︎    Home", "⏳    Filtering", "🕵    Filtered data", "👥   All data"]
        )

    def function(self):
        if self.page == "🏠︎    Home":
            self.home()
        elif self.page == "⏳    Filtering":
            self.filtering()
        elif self.page == "🕵    Filtered data":
            self.filtered_data()
        elif self.page == "👥   All data":
            self.all_data()

    def home(self):
        self.st.title("- - - -WELCOME TO 📊- - - -")
        self.st.title("PLACEMENT ELIGIBILITY WEB-APP")
        self.st.image("/content/illustrated-woman-being-intern-company.png")

    def filtering(self):
        self.st.title("ELIGIBLE STUDENTS 🎖️ ")
        self.st.sidebar.title("Enter the eligibility limits")
        codekata = self.st.sidebar.number_input("Codekata limit")
        assessments = self.st.sidebar.number_input("Assessments")
        projects = self.st.sidebar.number_input("Mini Projects")
        communication = self.st.sidebar.number_input("Communication score")
        mock_interview_score = self.st.sidebar.number_input("Mock interview score")

        if self.st.button("Run 🕵️"):
            query = '''
                SELECT
                    students_table.student_id,
                    students_table.name,
                    students_table.gender,
                    programming_table.problems_solve,
                    programming_table.assessments_complete,
                    programming_table.mini_projects,
                    soft_skills_table.communication,
                    placements_table.mock_interview_score
                FROM students_table
                INNER JOIN programming_table ON students_table.student_id = programming_table.student_id
                INNER JOIN soft_skills_table ON students_table.student_id = soft_skills_table.student_id
                INNER JOIN placements_table ON students_table.student_id = placements_table.student_id
                WHERE programming_table.problems_solve >= ?
                AND programming_table.assessments_complete >= ?
                AND programming_table.mini_projects >= ?
                AND soft_skills_table.communication >= ?
                AND placements_table.mock_interview_score >= ?
            '''

            df = self.mydb.execute_query(query, (codekata, assessments, projects, communication, mock_interview_score))
            self.st.dataframe(df)

            table = '''
                SELECT
                    students_table.student_id,
                    students_table.name,
                    students_table.age,
                    students_table.gender,
                    students_table.email,
                    students_table.phone_no,
                    students_table.enrollment_year,
                    students_table.course_batch,
                    students_table.city,
                    students_table.graduation_year,
                    programming_table.prog_language,
                    programming_table.problems_solve,
                    programming_table.assessments_complete,
                    programming_table.mini_projects,
                    programming_table.certificates_earn,
                    programming_table.latest_project_score,
                    soft_skills_table.communication,
                    soft_skills_table.teamwork,
                    soft_skills_table.presentation,
                    soft_skills_table.leadership,
                    soft_skills_table.criticalthinking,
                    soft_skills_table.interpersonal_skills,
                    placements_table.mock_interview_score,
                    placements_table.internships,
                    placements_table.placement_status,
                    placements_table.company_name,
                    placements_table.package,
                    placements_table.interview_rounds_cleared,
                    placements_table.placement_date
                FROM students_table
                INNER JOIN programming_table ON students_table.student_id = programming_table.student_id
                INNER JOIN soft_skills_table ON students_table.student_id = soft_skills_table.student_id
                INNER JOIN placements_table ON students_table.student_id = placements_table.student_id
                WHERE programming_table.problems_solve >= ?
                AND programming_table.assessments_complete >= ?
                AND programming_table.mini_projects >= ?
                AND soft_skills_table.communication >= ?
                AND placements_table.mock_interview_score >= ?
            '''

            df1 = self.mydb.execute_query(table, (codekata, assessments, projects, communication, mock_interview_score))
            self.mydb.save_dataframe(df1, 'filtered_table')

    def filtered_data(self):
        self.st.title("FILTERED DATA INSIGHTS 📊 ")
        queries_1 = {
            "Ranking codekata wise": '''
                SELECT RANK() OVER (ORDER BY problems_solve DESC) AS rank,
                       student_id, name, problems_solve
                FROM filtered_table
            ''',
            "Average soft skills score in each student": '''
                SELECT student_id, name,
                ROUND((communication + presentation + teamwork + leadership + criticalthinking + interpersonal_skills)/6.0, 2)
                AS Average_soft_skill_score
                FROM filtered_table
                WHERE (communication + presentation + teamwork + leadership + criticalthinking + interpersonal_skills)/6.0 > 60
            ''',
            "Count of students by placement status": '''
                SELECT placement_status, COUNT(*) AS total_students
                FROM filtered_table
                GROUP BY placement_status
            ''',
            "High score in mock interview and ready for job": '''
                SELECT student_id, name, mock_interview_score, internships, placement_status
                FROM filtered_table
                WHERE placement_status = 'ready'
                AND mock_interview_score > (SELECT AVG(mock_interview_score) FROM filtered_table)
                AND internships = 0
            ''',
            "Graduated more than 3 years but not placed": '''
                SELECT student_id, name, graduation_year, placement_status
                FROM filtered_table
                WHERE placement_status IN ('ready', 'not ready')
                AND graduation_year < (CAST(strftime('%Y', 'now') AS INTEGER) - 3)
                ORDER BY graduation_year
            '''
        }

        selected = self.st.selectbox("Choose a Query", list(queries_1.keys()))
        if self.st.button("Run"):
            df = self.mydb.execute_query(queries_1[selected])
            self.st.dataframe(df)

    def all_data(self):
        self.st.title("OVERALL DATA INSIGHTS  📊")
        queries_2 = {
            "Batch wise codekata average": '''
                SELECT students_table.course_batch, ROUND(AVG(programming_table.problems_solve)) AS average_codekata
                FROM students_table
                INNER JOIN programming_table ON students_table.student_id = programming_table.student_id
                GROUP BY students_table.course_batch
                ORDER BY students_table.course_batch
            ''',
            "Interview rounds cleared students counts": '''
                SELECT interview_rounds_cleared, COUNT(*) AS interview_rounds
                FROM placements_table
                GROUP BY interview_rounds_cleared
            ''',
            "Students placed  package above average": '''
                SELECT students_table.student_id,students_table.name,
                placements_table.company_name,placements_table.package
                FROM students_table INNER JOIN placements_table ON students_table.student_id =
                placements_table.student_id WHERE  placement_status = 'placed' AND CAST(REPLACE(placements_table.package, '₹', '') AS INTEGER) >
                ( SELECT AVG(CAST(REPLACE(package, '₹', '') AS INTEGER)) FROM placements_table WHERE placement_status = 'placed' )
                ORDER BY CAST(REPLACE(placements_table.package, '₹', '') AS INTEGER) DESC ''',
            "Top soft skill scores": '''
                SELECT students_table.student_id, students_table.name,
                ROUND((
                    soft_skills_table.communication +
                    soft_skills_table.teamwork +
                    soft_skills_table.presentation +
                    soft_skills_table.leadership +
                    soft_skills_table.criticalthinking +
                    soft_skills_table.interpersonal_skills
                ) / 6.0, 2) AS avg_soft_skill
                FROM students_table
                INNER JOIN soft_skills_table ON students_table.student_id = soft_skills_table.student_id
                ORDER BY avg_soft_skill DESC
                LIMIT 5
            ''',
            "Students placed with low average project score": '''
                SELECT students_table.student_id, students_table.name, programming_table.latest_project_score,
                       placements_table.placement_status
                FROM students_table
                INNER JOIN programming_table ON students_table.student_id = programming_table.student_id
                INNER JOIN placements_table ON students_table.student_id = placements_table.student_id
                WHERE placement_status = 'placed'
                AND programming_table.latest_project_score < (
                    SELECT AVG(latest_project_score) FROM programming_table
                )
                ORDER BY programming_table.latest_project_score
            '''
        }

        selected = self.st.selectbox("Choose a Query", list(queries_2.keys()))
        if self.st.button("Run 🕵️"):
            df = self.mydb.execute_query(queries_2[selected])
            self.st.dataframe(df)

run = StreamlitPage(pd, st)
run.function()
```
